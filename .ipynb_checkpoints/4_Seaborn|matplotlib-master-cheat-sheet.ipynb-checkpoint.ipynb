````python
Start:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

## 1. Import & Setup

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Standard Import** | `import matplotlib` | Import core matplotlib library | `import matplotlib.pyplot as plt` |
| **Seaborn Import** | `import seaborn` | Import seaborn with alias | `import seaborn as sns` |
| **Set Theme** | `sns.set_theme()` | Apply seaborn default theme/style | `sns.set_theme(style='whitegrid', palette='pastel')` |
| **Inline Magic** | `%matplotlib inline` | Display plots inline in Jupyter | `%matplotlib inline` |
| **Interactive Mode** | `plt.ion()` / `plt.ioff()` | Enable/disable interactive plotting | `plt.ion()` |
| **Backend** | `matplotlib.use()` | Set rendering backend | `matplotlib.use('Agg')` # for non-interactive |
| **Style Sheet** | `plt.style.use()` | Apply predefined style | `plt.style.use('seaborn-v0_8-darkgrid')` |
| **RC Params** | `plt.rcParams` | Global configuration dictionary | `plt.rcParams['figure.figsize'] = (10, 6)` |

**Available Styles:** `bmh`, `classic`, `dark_background`, `fast`, `fivethirtyeight`, `ggplot`, `grayscale`, `seaborn-v0_8`, `seaborn-v0_8-bright`, `seaborn-v0_8-colorblind`, `seaborn-v0_8-dark`, `seaborn-v0_8-darkgrid`, `seaborn-v0_8-deep`, `seaborn-v0_8-muted`, `seaborn-v0_8-notebook`, `seaborn-v0_8-paper`, `seaborn-v0_8-pastel`, `seaborn-v0_8-poster`, `seaborn-v0_8-talk`, `seaborn-v0_8-ticks`, `seaborn-v0_8-white`, `seaborn-v0_8-whitegrid`, `tableau-colorblind10`

---

## 2. Matplotlib.pyplot - Basic Plotting

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Line Plot** | `plt.plot()` | Draw lines and/or markers | `plt.plot(x, y, 'ro--', label='Line 1')` |
| **Scatter Plot** | `plt.scatter()` | Create scatter plot | `plt.scatter(x, y, c=z, s=100, cmap='viridis', alpha=0.7)` |
| ** --> Bar Chart (Vertical)** | `plt.bar()` | Vertical bar chart | `plt.bar(top10.index, top10.values, color="steelblue", edgecolor="black") plt.xlabel("asdf") plt.ylabel("Frequency") plt.xticks(rotation=45) plt.show()` |
| **Bar Chart (Horizontal)** | `plt.barh()` | Horizontal bar chart | `plt.barh(y, width, color='coral')` |
| **Histogram** | `plt.hist()` | Frequency distribution | `plt.hist(data, bins=30, density=True, alpha=0.7)` |
| **Box Plot** | `plt.boxplot()` | Statistical summary box | `plt.boxplot(data, vert=False, patch_artist=True)` |
| **Pie Chart** | `plt.pie()` | Circular proportion chart | `plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)` |
| **Area Fill** | `plt.fill_between()` | Fill area between curves | `plt.fill_between(x, y1, y2, alpha=0.3, color='green')` |
| **Stacked Area** | `plt.stackplot()` | Stacked area chart | `plt.stackplot(x, y1, y2, y3, labels=['A','B','C'])` |
| **Step Plot** | `plt.step()` | Step function plot | `plt.step(x, y, where='mid', color='purple')` |
| **Error Bars** | `plt.errorbar()` | Plot with error bars | `plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5)` |
| **Stem Plot** | `plt.stem()` | Discrete sequence plot | `plt.stem(x, y, linefmt='r-', markerfmt='ro')` |
| **Violin (basic)** | `plt.violinplot()` | Violin distribution plot | `plt.violinplot(data, showmeans=True, showmedians=True)` |
| **Polar Plot** | `plt.polar()` | Polar coordinate plot | `plt.polar(theta, r, 'o-')` |
| **Log Plot** | `plt.loglog()` / `plt.semilogx()` / `plt.semilogy()` | Logarithmic scale plots | `plt.semilogx(x, y)` |
| **Event Plot** | `plt.eventplot()` | Parallel events/lines | `plt.eventplot(data, orientation='horizontal', colors='black')` |

### Plot Format Strings (fmt)
```python
# Format: '[marker][line][color]'
plt.plot(x, y, 'ro--')  # red circles, dashed line
# Markers: . o v ^ < > s p * h H + x D d | _
# Lines: - -- -. :
# Colors: b g r c m y k w (or hex: '#FF5733')
```

---

## 3. Matplotlib.pyplot - Customization

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Title** | `plt.title()` | Add plot title | `plt.title('My Plot', fontsize=14, fontweight='bold', pad=20)` |
| **X Label** | `plt.xlabel()` | X-axis label | `plt.xlabel('Time (s)', fontsize=12, color='navy')` |
| **Y Label** | `plt.ylabel()` | Y-axis label | `plt.ylabel('Amplitude', fontsize=12)` |
| **Legend** | `plt.legend()` | Add legend | `plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=2)` |
| **Grid** | `plt.grid()` | Add grid lines | `plt.grid(True, linestyle='--', alpha=0.7, color='gray')` |
| **X Ticks** | `plt.xticks()` | Set x-axis ticks | `plt.xticks([0, 1, 2], ['A', 'B', 'C'], rotation=45)` |
| **Y Ticks** | `plt.yticks()` | Set y-axis ticks | `plt.yticks(range(0, 101, 10))` |
| **X Limits** | `plt.xlim()` | Set x-axis range | `plt.xlim(0, 100)` |
| **Y Limits** | `plt.ylim()` | Set y-axis range | `plt.ylim(-5, 5)` |
| **Axis Limits** | `plt.axis()` | Set all axis limits | `plt.axis([xmin, xmax, ymin, ymax])` |
| **Tick Parameters** | `plt.tick_params()` | Customize ticks appearance | `plt.tick_params(axis='both', labelsize=10, colors='red')` |
| **Tick Labels** | `plt.set_xticklabels()` | Set custom tick labels (OO) | `ax.set_xticklabels(['Jan','Feb','Mar'])` |
| **Spines** | `ax.spines` | Control border lines (OO) | `ax.spines['top'].set_visible(False)` |
| **Colorbar** | `plt.colorbar()` | Add color scale bar | `plt.colorbar(mappable, label='Intensity', shrink=0.8)` |
| **Clim** | `plt.clim()` | Set color limits | `plt.clim(vmin=0, vmax=100)` |
| **Margins** | `plt.margins()` | Add padding around data | `plt.margins(x=0.1, y=0.1)` |
| **Autoscale** | `plt.autoscale()` | Auto-adjust axis limits | `plt.autoscale(enable=True, axis='both', tight=True)` |

### Legend Location Codes
```python
# Location strings: 'best', 'upper right', 'upper left', 'lower left', 'lower right'
# 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
# Or use numeric codes: 0='best', 1='upper right', 2='upper left', etc.
```

---

## 4. Matplotlib.pyplot - Grid & Layout

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Create Figure** | `plt.figure()` | Create new figure | `fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='white')` |
| **Subplots** | `plt.subplots()` | Create figure + axes grid | `fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharex=True)` |
| **Subplot** | `plt.subplot()` | Add single subplot | `plt.subplot(2, 2, 1)` # or `plt.subplot(221)` |
| **Subplot Mosaic** | `plt.subplot_mosaic()` | Flexible layout (≥3.6) | `fig, ax_dict = plt.subplot_mosaic([['A', 'B'], ['C', 'C']])` |
| **Add Subplot** | `fig.add_subplot()` | Add subplot to figure (OO) | `ax = fig.add_subplot(2, 2, 1)` |
| **Add Axes** | `fig.add_axes()` | Add axes at specific position | `ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])` |
| **GridSpec** | `plt.GridSpec()` | Complex grid layout | `gs = fig.add_gridspec(3, 3); ax = fig.add_subplot(gs[0, :])` |
| **Tight Layout** | `plt.tight_layout()` | Auto-adjust spacing | `plt.tight_layout(pad=2.0, h_pad=1.0, w_pad=1.0)` |
| **Constrained Layout** | `plt.subplots(..., layout='constrained')` | Better auto-spacing | `fig, ax = plt.subplots(layout='constrained')` |
| **Subplots Adjust** | `plt.subplots_adjust()` | Manual spacing adjustment | `plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.2, hspace=0.3)` |
| **Figure Size** | `fig.set_size_inches()` | Change figure size (OO) | `fig.set_size_inches(10, 6)` |
| **Figure DPI** | `fig.set_dpi()` | Set resolution (OO) | `fig.set_dpi(300)` |
| **Supertitle** | `fig.suptitle()` | Overall figure title (OO) | `fig.suptitle('Figure Title', fontsize=16, y=1.02)` |
| **Get Current Figure** | `plt.gcf()` | Get current figure object | `fig = plt.gcf()` |
| **Get Current Axes** | `plt.gca()` | Get current axes object | `ax = plt.gca()` |
| **Twin X** | `ax.twinx()` | Secondary y-axis (OO) | `ax2 = ax.twinx(); ax2.plot(x, y2, 'r-')` |
| **Twin Y** | `ax.twiny()` | Secondary x-axis (OO) | `ax2 = ax.twiny()` |
| **Zoom/Pan** | `plt.xlim()` / `plt.ylim()` | Interactive zooming | Dynamic adjustment |

---

## 5. Matplotlib.pyplot - Text & Annotations

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Text** | `plt.text()` | Add text at coordinates | `plt.text(0.5, 0.5, 'Hello', fontsize=12, ha='center')` |
| **Annotate** | `plt.annotate()` | Annotate with arrow | `plt.annotate('Peak', xy=(2, 10), xytext=(3, 12), arrowprops=dict(arrowstyle='->', color='red'))` |
| **Arrow** | `plt.arrow()` | Draw arrow | `plt.arrow(0, 0, 1, 1, head_width=0.1, color='blue')` |
| **Vertical Line** | `plt.axvline()` | Vertical reference line | `plt.axvline(x=5, color='red', linestyle='--', linewidth=2, label='Threshold')` |
| **Horizontal Line** | `plt.axhline()` | Horizontal reference line | `plt.axhline(y=0, color='black', linewidth=0.5)` |
| **Vertical Span** | `plt.axvspan()` | Vertical shaded region | `plt.axvspan(xmin=2, xmax=4, alpha=0.2, color='yellow')` |
| **Horizontal Span** | `plt.axhspan()` | Horizontal shaded region | `plt.axhspan(ymin=0, ymax=1, alpha=0.3, color='green')` |
| **XLine** | `plt.axline()` | Infinite line (≥3.3) | `plt.axline((0, 0), slope=1, color='black', linestyle='--')` |
| **FigText** | `fig.text()` | Text in figure coordinates (OO) | `fig.text(0.5, 0.02, 'Caption', ha='center')` |
| **Table** | `plt.table()` | Add table to plot | `plt.table(cellText=data, colLabels=columns, loc='bottom')` |
| **Math Text** | `r'$...$'` | LaTeX-style math rendering | `plt.title(r'$\\alpha > \\beta$')` |
| **Font Properties** | `matplotlib.font_manager` | Custom fonts | `from matplotlib.font_manager import FontProperties; fp = FontProperties(family='serif')` |

---

## 6. Matplotlib.pyplot - Advanced Features

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Contour Plot** | `plt.contour()` | Contour lines | `plt.contour(X, Y, Z, levels=10, colors='black')` |
| **Filled Contour** | `plt.contourf()` | Filled contour plot | `plt.contourf(X, Y, Z, levels=20, cmap='RdYlBu')` |
| **Imshow** | `plt.imshow()` | Display 2D array as image | `plt.imshow(data, cmap='hot', interpolation='nearest', aspect='auto')` |
| **Pcolormesh** | `plt.pcolormesh()` | Pseudocolor mesh plot | `plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')` |
| **Quiver** | `plt.quiver()` | Vector field arrows | `plt.quiver(X, Y, U, V, color='red')` |
| **Streamplot** | `plt.streamplot()` | Streamline plot | `plt.streamplot(X, Y, U, V, density=2, color='blue')` |
| **Triplot** | `plt.triplot()` | Triangular grid plot | `plt.triplot(x, y, triangles, 'go-')` |
| **Tricontour** | `plt.tricontour()` | Contour on triangulation | `plt.tricontour(x, y, z, levels=10)` |
| **Tricontourf** | `plt.tricontourf()` | Filled contour on triangulation | `plt.tricontourf(x, y, z, levels=15)` |
| **Hexbin** | `plt.hexbin()` | Hexagonal binning | `plt.hexbin(x, y, gridsize=20, cmap='Blues', mincnt=1)` |
| **Specgram** | `plt.specgram()` | Spectrogram | `plt.specgram(x, NFFT=256, Fs=2, noverlap=128)` |
| **XCorr** | `plt.xcorr()` | Cross-correlation plot | `plt.xcorr(x, y, maxlags=50)` |
| **ACorr** | `plt.acorr()` | Auto-correlation plot | `plt.acorr(x, maxlags=50)` |
| **Boxplot (enhanced)** | `plt.bxp()` | Draw boxplots from dict | `plt.bxp(stats_dict, showfliers=True)` |
| **Broken Barh** | `plt.broken_barh()` | Horizontal broken bar | `plt.broken_barh([(10, 50), (100, 30)], (10, 9), facecolors='tab:blue')` |
| **Matshow** | `plt.matshow()` | Matrix display in new figure | `plt.matshow(data, cmap='coolwarm')` |
| **Spy** | `plt.spy()` | Sparse matrix visualization | `plt.spy(matrix, markersize=5, color='black')` |
| **3D Projection** | `fig.add_subplot(projection='3d')` | 3D plotting setup | `ax = fig.add_subplot(111, projection='3d'); ax.plot3D(x, y, z)` |
| **Image Read** | `matplotlib.image.imread()` | Load image from file | `img = plt.imread('image.png')` |
| **Image Show** | `plt.imsave()` / `plt.imshow()` | Save/show image array | `plt.imsave('output.png', img_array)` |

---

## 7. Matplotlib.pyplot - Output & Saving

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Show Plot** | `plt.show()` | Display the figure | `plt.show()` |
| **Save Figure** | `plt.savefig()` | Save to file | `plt.savefig('plot.png', dpi=300, bbox_inches='tight', transparent=True)` |
| **Clear Figure** | `plt.clf()` | Clear current figure | `plt.clf()` |
| **Clear Axes** | `plt.cla()` | Clear current axes | `plt.cla()` |
| **Close Figure** | `plt.close()` | Close figure window | `plt.close('all')` # close all figures |
| **Draw** | `plt.draw()` | Redraw the figure | `plt.draw()` |
| **Pause** | `plt.pause()` | Pause for animation | `plt.pause(0.1)` |
| **Connect** | `plt.connect()` | Connect event handler | `plt.connect('button_press_event', onclick)` |
| **Disconnect** | `plt.disconnect()` | Disconnect event handler | `plt.disconnect(cid)` |
| **GInput** | `plt.ginput()` | Get mouse clicks | `points = plt.ginput(n=3, timeout=30)` |
| **Waitforbuttonpress** | `plt.waitforbuttonpress()` | Wait for key/mouse | `plt.waitforbuttonpress(timeout=-1)` |

### Savefig Parameters
```python
plt.savefig('figure.pdf', 
            dpi=300,           # Resolution
            format='pdf',      # File format: png, pdf, svg, eps, jpg, etc.
            bbox_inches='tight',  # Remove extra whitespace
            pad_inches=0.1,    # Padding around figure
            transparent=True,  # Transparent background
            facecolor='white', # Figure face color
            edgecolor='none')  # Figure edge color
```

---

## 8. Seaborn - Figure-Level Functions

> **Figure-level functions** create their own figure and return a `FacetGrid`, `PairGrid`, or `JointGrid` object. They support faceting via `col` and `row` parameters and use `height`/`aspect` for sizing.

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Relational Plot** | `sns.relplot()` | Figure-level relational plots | `sns.relplot(data=df, x='x', y='y', hue='group', col='category', kind='scatter', height=4, aspect=1.2)` |
| **Distribution Plot** | `sns.displot()` | Figure-level distribution plots | `sns.displot(data=df, x='value', hue='group', kind='hist', kde=True, col='category', height=3)` |
| **Categorical Plot** | `sns.catplot()` | Figure-level categorical plots | `sns.catplot(data=df, x='cat', y='val', hue='group', kind='box', col='subcat', height=4)` |

### Figure-Level Parameters (Common)
```python
# data: DataFrame
# x, y: Column names for axes
# hue: Color grouping variable
# col: Facet across columns
# row: Facet across rows
# kind: Plot type ('scatter', 'line', 'hist', 'kde', 'ecdf', 'strip', 'swarm', 'box', 'violin', 'boxen', 'point', 'bar', 'count')
# height: Height of each facet (inches)
# aspect: Width = height * aspect
# palette: Color palette name
# legend: Legend control ('auto', 'brief', 'full', False)
# margin_titles: Put titles on margins
# facet_kws: Additional FacetGrid parameters
```

---

## 9. Seaborn - Relational Plots (Axes-Level)

> **Axes-level functions** plot onto a single `matplotlib.axes.Axes` and can be composed into complex figures using the `ax=` parameter.

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Scatter Plot** | `sns.scatterplot()` | Scatter plot with styling | `sns.scatterplot(data=df, x='x', y='y', hue='group', size='size_col', style='style_col', palette='deep', alpha=0.7)` |
| **Line Plot** | `sns.lineplot()` | Line plot with confidence interval | `sns.lineplot(data=df, x='x', y='y', hue='group', ci='sd', estimator='mean', err_style='band')` |

### Relational Plot Parameters
```python
# data, x, y, hue, size, style: Core mapping variables
# palette: Color palette
# hue_order: Order for hue categories
# hue_norm: Normalize hue for numeric
# sizes: Size range (min, max)
# size_norm: Normalize size
# markers: Marker styles dict or list
# style_order: Order for style categories
# legend: Legend mode ('auto', 'brief', 'full', False)
# ax: Matplotlib axes to plot on
# alpha: Transparency
```

---

## 10. Seaborn - Distribution Plots (Axes-Level)

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Histogram** | `sns.histplot()` | Histogram with bins | `sns.histplot(data=df, x='value', hue='group', bins=30, kde=True, stat='density', multiple='stack', element='bars')` |
| **KDE Plot** | `sns.kdeplot()` | Kernel density estimate | `sns.kdeplot(data=df, x='value', hue='group', fill=True, bw_adjust=1.5, cut=0, cumulative=False)` |
| **ECDF Plot** | `sns.ecdfplot()` | Empirical CDF | `sns.ecdfplot(data=df, x='value', hue='group', stat='proportion', complementary=False)` |
| **Rug Plot** | `sns.rugplot()` | Marginal rug marks | `sns.rugplot(data=df, x='value', hue='group', height=0.05, expand_margins=True)` |
| **Deprecated** | `sns.distplot()` | **DEPRECATED** - Use displot/histplot | `sns.histplot(data=df, x='value', kde=True)` # replacement |

### Distribution Plot Parameters
```python
# data, x, y: Variables to plot
# hue: Grouping variable
# stat: Statistic ('count', 'frequency', 'density', 'probability', 'proportion', 'percent')
# bins: Number of bins or bin specification
# binwidth: Width of each bin
# binrange: Range for bins (min, max)
# kde: Overlay KDE curve (histplot only)
# fill: Fill under curve (kdeplot)
# common_norm: Normalize across all hue groups
# common_grid: Use same evaluation grid
# cumulative: Plot cumulative distribution
# multiple: How to handle multiple groups ('layer', 'dodge', 'stack', 'fill')
# element: Histogram element ('bars', 'step', 'poly')
# shrink: Shrink bars relative to binwidth
# log_scale: Log scale for axis
```

---

## 11. Seaborn - Categorical Plots (Axes-Level)

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Strip Plot** | `sns.stripplot()` | Scatter plot with jitter | `sns.stripplot(data=df, x='cat', y='val', hue='group', jitter=True, dodge=True, alpha=0.6)` |
| **Swarm Plot** | `sns.swarmplot()` | Non-overlapping scatter | `sns.swarmplot(data=df, x='cat', y='val', hue='group', dodge=True, size=4)` |
| **Box Plot** | `sns.boxplot()` | Box-and-whisker plot | `sns.boxplot(data=df, x='cat', y='val', hue='group', whis=1.5, showfliers=True, notch=False)` |
| **Violin Plot** | `sns.violinplot()` | Violin distribution plot | `sns.violinplot(data=df, x='cat', y='val', hue='group', split=True, inner='box', bw=0.2)` |
| **Boxen Plot** | `sns.boxenplot()` | Letter-value plot (enhanced box) | `sns.boxenplot(data=df, x='cat', y='val', hue='group', k_depth='proportion')` |
| **Point Plot** | `sns.pointplot()` | Point estimate with CI | `sns.pointplot(data=df, x='cat', y='val', hue='group', ci='sd', capsize=0.1, join=True)` |
| **Bar Plot** | `sns.barplot()` | Bar with confidence interval | `sns.barplot(data=df, x='cat', y='val', hue='group', ci='sd', errcolor='gray', errwidth=2)` |
| **Count Plot** | `sns.countplot()` | Count of observations | `sns.countplot(data=df, x='cat', hue='group', dodge=True)` |

### Categorical Plot Parameters
```python
# data, x, y: Variables (one must be categorical)
# hue: Grouping variable
# order: Order for categorical variable
# hue_order: Order for hue categories
# estimator: Aggregation function (barplot/pointplot): 'mean', 'median', sum, etc.
# ci: Confidence interval ('sd', int for percentile, None)
# n_boot: Number of bootstrap iterations
# units: Unit identifier for repeated measures
# seed: Random seed for bootstrap
# orient: Orientation ('v', 'h')
# color: Single color
# palette: Color palette
# saturation: Saturation of colors
# errcolor: Error bar color
# errwidth: Error bar width
# capsize: Cap size for error bars
# dodge: Separate hue levels
# ax: Matplotlib axes
```

---

## 12. Seaborn - Regression Plots

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Reg Plot** | `sns.regplot()` | Scatter + regression line (axes-level) | `sns.regplot(data=df, x='x', y='y', scatter=True, ci=95, order=1, robust=False, logx=False)` |
| **LM Plot** | `sns.lmplot()` | Regression plot with facets (figure-level) | `sns.lmplot(data=df, x='x', y='y', hue='group', col='cat', row='subcat', ci=95, height=4, aspect=1)` |
| **Resid Plot** | `sns.residplot()` | Residuals of regression | `sns.residplot(data=df, x='x', y='y', lowess=True, scatter_kws={'alpha':0.5})` |

### Regression Parameters
```python
# data, x, y: Variables
# hue: Grouping variable (lmplot only)
# col, row: Faceting variables (lmplot only)
# order: Polynomial order (1=linear, 2=quadratic)
# logx: Fit to log(x)
# logistic: Logistic regression
# lowess: Locally weighted regression
# robust: Robust regression (statsmodels)
# ci: Confidence interval (int or None)
# n_boot: Bootstrap iterations
# scatter: Show scatter points
# fit_reg: Show regression line
# truncate: Truncate regression at data limits
# x_jitter, y_jitter: Add jitter
# label: Label for legend
# color: Single color
# marker: Marker style
# scatter_kws: Dict for scatter plot params
# line_kws: Dict for line params
# ax: Axes (regplot only)
```

---

## 13. Seaborn - Matrix & Cluster Plots

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Heatmap** | `sns.heatmap()` | Color-coded matrix | `sns.heatmap(data, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5, linecolor='white', cbar_kws={'label': 'Value'}, vmin=-1, vmax=1, center=0, square=True, mask=mask)` |
| **Clustermap** | `sns.clustermap()` | Hierarchical clustered heatmap | `sns.clustermap(data, cmap='viridis', standard_scale=1, figsize=(10, 8), dendrogram_ratio=0.2, cbar_pos=(0.02, 0.8, 0.05, 0.18), row_cluster=True, col_cluster=True, method='average', metric='euclidean')` |

### Heatmap Parameters
```python
# data: 2D dataset
# vmin, vmax: Color scale limits
# cmap: Colormap name
# center: Center value for diverging colormap
# robust: Use robust quantiles for color limits
# annot: Annotate cells with values
# fmt: Format string for annotations
# annot_kws: Keyword args for annotations
# linewidths: Width of cell borders
# linecolor: Color of cell borders
# cbar: Show colorbar
# cbar_kws: Colorbar keyword arguments
# cbar_ax: Axes for colorbar
# square: Make cells square
# xticklabels, yticklabels: Label control
# mask: Boolean mask for hiding cells
# ax: Matplotlib axes
```

### Clustermap Parameters
```python
# data: 2D dataset
# pivot_kws: Pivot parameters
# method: Linkage method ('single', 'complete', 'average', 'weighted', 'centroid', 'median', 'ward')
# metric: Distance metric ('euclidean', 'cityblock', 'cosine', etc.)
# z_score: Z-score normalization (0=rows, 1=cols)
# standard_scale: Standard scale (0=rows, 1=cols)
# figsize: Figure size
# cbar_kws: Colorbar parameters
# row_cluster, col_cluster: Enable clustering
# row_linkage, col_linkage: Pre-computed linkage
# row_colors, col_colors: Color labels for rows/cols
# tree_kws: Dendrogram parameters
# dendrogram_ratio: Ratio of dendrogram to heatmap
# cbar_pos: Colorbar position (x, y, width, height)
# mask: Boolean mask
```

---

## 14. Seaborn - Multi-Plot Grids

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Facet Grid** | `sns.FacetGrid()` | Grid of plots by variables | `g = sns.FacetGrid(data=df, col='cat', row='subcat', hue='group', height=3, aspect=1.2, margin_titles=True); g.map_dataframe(sns.scatterplot, x='x', y='y'); g.add_legend()` |
| **Pair Grid** | `sns.PairGrid()` | Grid of pairwise relationships | `g = sns.PairGrid(data=df, hue='group', vars=['a','b','c'], corner=True, diag_sharey=False); g.map_diag(sns.histplot); g.map_offdiag(sns.scatterplot); g.add_legend()` |
| **Pair Plot** | `sns.pairplot()` | Convenience pair grid (figure-level) | `sns.pairplot(data=df, hue='group', vars=['a','b','c'], diag_kind='hist', corner=True, plot_kws={'alpha':0.6})` |
| **Joint Grid** | `sns.JointGrid()` | Joint + marginal plots | `g = sns.JointGrid(data=df, x='x', y='y', height=6, ratio=4, space=0.2); g.plot_joint(sns.scatterplot, hue='group'); g.plot_marginals(sns.histplot, kde=True)` |
| **Joint Plot** | `sns.jointplot()` | Convenience joint grid (figure-level) | `sns.jointplot(data=df, x='x', y='y', hue='group', kind='scatter', height=6, ratio=4, marginal_ticks=True)` |

### FacetGrid Methods
```python
g.map(func, *args, **kwargs)        # Map function to each facet
g.map_dataframe(func, *args, **kwargs)  # Map with DataFrame access
g.map_diag(func, **kwargs)          # Map to diagonal (PairGrid)
g.map_offdiag(func, **kwargs)       # Map to off-diagonal (PairGrid)
g.map_lower(func, **kwargs)         # Map to lower triangle
g.map_upper(func, **kwargs)         # Map to upper triangle
g.set_axis_labels(x_var, y_var)     # Set axis labels
g.set_titles(template)              # Set subplot titles
g.set(xlim=..., ylim=...)          # Set axis limits
g.add_legend()                      # Add legend
g.fig.suptitle('Title')             # Figure title
g.savefig('file.png')               # Save figure
```

### Joint Plot Kinds
```python
kind='scatter'   # Scatter + histograms/kde
kind='hist'      # Hexbin/2D histogram + histograms
kind='kde'       # KDE contour + KDE marginals
kind='reg'       # Regression + histograms
kind='resid'     # Residuals + histograms
```

---

## 15. Seaborn - Themes & Palettes

| Theme | Query | Explanation / Keywords | Syntax Example |
|-------|-------|------------------------|----------------|
| **Set Theme** | `sns.set_theme()` | Set overall theme | `sns.set_theme(style='whitegrid', context='notebook', palette='deep', font='sans-serif', font_scale=1)` |
| **Set Style** | `sns.set_style()` | Set aesthetic style | `sns.set_style('darkgrid')` # white, dark, whitegrid, darkgrid, ticks |
| **Set Context** | `sns.set_context()` | Set plotting context | `sns.set_context('talk')` # paper, notebook, talk, poster |
| **Set Palette** | `sns.set_palette()` | Set color palette | `sns.set_palette('husl', n_colors=10)` |
| **Color Palette** | `sns.color_palette()` | Get/create color palette | `palette = sns.color_palette('coolwarm', n_colors=7, as_cmap=True)` |
| **HLS Palette** | `sns.hls_palette()` | HLS color palette | `sns.hls_palette(n_colors=8, l=0.5, s=0.8)` |
| **Husl Palette** | `sns.husl_palette()` | HUSL color palette | `sns.husl_palette(n_colors=8, h=0.5, s=0.8, l=0.6)` |
| **Cubehelix** | `sns.cubehelix_palette()` | Cubehelix palette | `sns.cubehelix_palette(n_colors=8, start=0.5, rot=-0.75, dark=0.3, light=0.8)` |
| **Diverging** | `sns.diverging_palette()` | Diverging palette | `sns.diverging_palette(250, 15, s=75, l=40, n=9, center='light')` |
| **Blends** | `sns.blend_palette()` | Blend between colors | `sns.blend_palette(['red', 'white', 'blue'], n_colors=9, as_cmap=True)` |
| **XKCD Colors** | `sns.xkcd_palette()` | XKCD named colors | `sns.xkcd_palette(['red', 'green', 'blue'])` |
| **Crayon Colors** | `sns.crayon_palette()` | Crayola crayon colors | `sns.crayon_palette(['Red', 'Blue', 'Green'])` |
| **Light Palette** | `sns.light_palette()` | Light sequential palette | `sns.light_palette('seagreen', n_colors=8, reverse=False)` |
| **Dark Palette** | `sns.dark_palette()` | Dark sequential palette | `sns.dark_palette('purple', n_colors=8, reverse=False)` |
| **Desaturate** | `sns.desaturate()` | Desaturate a color | `sns.desaturate('red', prop=0.5)` |
| **Saturate** | `sns.saturate()` | Saturate a color | `sns.saturate('red', prop=0.5)` |

### Built-in Palettes
```python
# Qualitative: 'deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind', 'Set1', 'Set2', 'Set3', 'Paired', 'tab10', 'tab20'
# Sequential: 'Blues', 'BuGn', 'BuPu', 'GnBu', 'Greens', 'Greys', 'Oranges', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'Purples', 'RdPu', 'Reds', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd', 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'rocket', 'mako', 'flare', 'crest'
# Diverging: 'RdBu', 'RdGy', 'PuOr', 'Spectral', 'coolwarm', 'bwr', 'seismic', 'vlag', 'icefire'
```

### Context Parameters
```python
# paper:   smallest fonts, thin lines
# notebook: default
# talk:     larger fonts, thicker lines
# poster:   largest fonts, thickest lines
# Parameters: font_scale, rc={'axes.labelsize': 14, ...}
```

---

## 16. Quick Reference: Matplotlib ↔ Seaborn Mapping

| Visualization | Matplotlib | Seaborn (Axes-Level) | Seaborn (Figure-Level) |
|---------------|------------|----------------------|------------------------|
| **Line Chart** | `plt.plot()` | `sns.lineplot()` | `sns.relplot(kind='line')` |
| **Scatter Plot** | `plt.scatter()` | `sns.scatterplot()` | `sns.relplot(kind='scatter')` |
| **Bar Chart** | `plt.bar()` / `plt.barh()` | `sns.barplot()` | `sns.catplot(kind='bar')` |
| **Histogram** | `plt.hist()` | `sns.histplot()` | `sns.displot(kind='hist')` |
| **KDE/Density** | `scipy.stats.gaussian_kde` | `sns.kdeplot()` | `sns.displot(kind='kde')` |
| **Box Plot** | `plt.boxplot()` | `sns.boxplot()` | `sns.catplot(kind='box')` |
| **Violin Plot** | `plt.violinplot()` | `sns.violinplot()` | `sns.catplot(kind='violin')` |
| **Heatmap** | `plt.imshow()` | `sns.heatmap()` | — |
| **Pairwise Grid** | Custom subplots | `sns.PairGrid()` | `sns.pairplot()` |
| **Joint Plot** | Custom subplots | `sns.JointGrid()` | `sns.jointplot()` |
| **Regression** | `np.polyfit` + `plt.plot()` | `sns.regplot()` | `sns.lmplot()` |
| **ECDF** | Custom cumulative | `sns.ecdfplot()` | `sns.displot(kind='ecdf')` |
| **Count/Frequency** | `plt.bar()` + `value_counts()` | `sns.countplot()` | `sns.catplot(kind='count')` |
| **Strip Plot** | `plt.scatter()` + jitter | `sns.stripplot()` | `sns.catplot(kind='strip')` |
| **Swarm Plot** | Custom | `sns.swarmplot()` | `sns.catplot(kind='swarm')` |
| **Point Plot** | Custom error bars | `sns.pointplot()` | `sns.catplot(kind='point')` |
| **Boxen/Letter** | Custom | `sns.boxenplot()` | `sns.catplot(kind='boxen')` |
| **Area Fill** | `plt.fill_between()` | — | — |
| **Pie Chart** | `plt.pie()` | — | — |
| **Contour** | `plt.contour()` / `plt.contourf()` | — | — |
| **Quiver** | `plt.quiver()` | — | — |
| **Hexbin** | `plt.hexbin()` | — | — |

---

## 🏷️ Common Parameters Quick Reference

### Universal Seaborn Parameters
```python
# data: pandas DataFrame
# x, y: Column names (strings)
# hue: Grouping variable for color
# size: Grouping variable for size
# style: Grouping variable for marker style
# palette: Color palette name or dict
# hue_order: List specifying hue category order
# hue_norm: Normalize hue for numeric (tuple or Normalize object)
# color: Single color for all elements
# ax: Matplotlib Axes object to draw on
# legend: 'auto', 'brief', 'full', or False
```

### Universal Matplotlib Parameters
```python
# color / c: Color (name, hex, RGB tuple)
# alpha: Transparency (0-1)
# linewidth / lw: Line width
# linestyle / ls: Line style ('-', '--', '-.', ':', '')
# marker: Marker style ('o', 's', '^', 'D', '*', etc.)
# markersize / ms: Marker size
# markerfacecolor / mfc: Marker fill color
# markeredgecolor / mec: Marker edge color
# markeredgewidth / mew: Marker edge width
# label: Label for legend
# zorder: Drawing order (higher = on top)
# visible: Show/hide element
# rasterized: Rasterize for PDF/SVG output
```

---

## 💡 Pro Tips

1. **Use Object-Oriented style for complex figures:** `fig, ax = plt.subplots()` gives you full control.
2. **Use pyplot style for quick EDA:** `plt.plot()`, `plt.scatter()` are faster for exploration.
3. **Figure-level vs Axes-level:** Use figure-level (`relplot`, `displot`, `catplot`) for faceting; use axes-level for composing complex multi-plot figures.
4. **Color palettes:** Use `sns.color_palette()` to preview palettes: `sns.palplot(sns.color_palette('viridis', 10))`
5. **Save high-quality figures:** Always use `bbox_inches='tight'` and `dpi=300` for publications.
6. **Memory management:** Call `plt.close('all')` after saving in loops to prevent memory leaks.
7. **Jupyter notebooks:** Use `%matplotlib inline` for static, `%matplotlib widget` for interactive.
8. **Seaborn + Matplotlib combo:** Seaborn axes-level functions return matplotlib Axes, so you can chain matplotlib methods: `sns.boxplot(...).set_title('Title')`
````

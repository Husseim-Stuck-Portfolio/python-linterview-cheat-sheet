```python
# Ultimate Python Statistics Cheat Sheet

#A comprehensive reference for data analysis with Pandas, NumPy, Matplotlib, and Seaborn. 
#Covers Categorical vs Categorical, Categorical vs Continuous, and Continuous vs Continuous relationships.

---

## IMPORTS
from scipy.stats import chi2_contingency
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.graphics.mosaicplot import mosaic
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

---

## CATEGORICAL VS CATEGORICAL

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Crosstab | pd.crosstab() | Frequency table with margins and normalization | pd.crosstab(df['Gender'], df['Product'], margins=True, normalize='all') |
| Crosstab Row Norm | pd.crosstab(normalize='index') | Proportions within each row category | pd.crosstab(df['Gender'], df['Product'], normalize='index') |
| Crosstab Col Norm | pd.crosstab(normalize='columns') | Proportions within each column category | pd.crosstab(df['Gender'], df['Product'], normalize='columns') |
| Crosstab Aggregate | pd.crosstab(aggfunc=) | Aggregate values by categories | pd.crosstab(df['Gender'], df['Product'], values=df['Sales'], aggfunc='mean') |
| Stacked Bar Charts | crosstab_result.plot(kind="bar", stacked=True) | Stacked and Grouped Bar Charts | crosstab_result.plot(kind="bar", stacked=True) |
| Chi-Square Test | scipy.stats.chi2_contingency | Test independence between two categorical variables | chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['A'], df['B'])) | chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result) |
| Cramer's V | Manual calculation | Association strength (0=none, 1=perfect) | n = cont.sum().sum(); phi2 = chi2/n; r, k = cont.shape; phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1)); rcorr = r-((r-1)**2)/(n-1); kcorr = k-((k-1)**2)/(n-1); v = np.sqrt(phi2corr/min((kcorr-1),(rcorr-1))) |
| Stacked Bar | df.plot(kind='bar', stacked=True) | Composition of categories | pd.crosstab(df['Gender'], df['Product']).plot(kind='bar', stacked=True, colormap='Set2') |
| Grouped Bar | df.plot(kind='bar') | Side-by-side comparison | pd.crosstab(df['Gender'], df['Product']).plot(kind='bar', color=['skyblue','salmon']) |
| Frequency Heatmap | sns.heatmap() | Color-coded frequency matrix | sns.heatmap(pd.crosstab(df['Gender'], df['Product']), annot=True, fmt='d', cmap='YlOrRd') |
| Mosaic Plot | statsmodels.graphics.mosaicplot | Proportional tile visualization | from statsmodels.graphics.mosaicplot import mosaic; mosaic(df, ['Gender','Product']) |

INTERPRETATION - Categorical vs Categorical:
- Chi-Square p < 0.05: Variables are statistically associated (reject independence)
- Chi-Square p >= 0.05: No significant association detected
- Cramer's V ~ 0: No association
- Cramer's V ~ 0.1: Weak association
- Cramer's V ~ 0.3: Moderate association
- Cramer's V ~ 0.5+: Strong association
- Unequal bar heights: Visual evidence of association between categories

---

## CATEGORICAL VS CONTINUOUS

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Group Stats | df.groupby().agg() | Aggregate by category | df.groupby('Gender')['Salary'].agg(['mean','std','count','min','max']) |
| Group Describe | df.groupby().describe() | Full descriptive stats per group | df.groupby('Gender')['Salary'].describe() |
| Violin Plot | sns.violinplot() | Distribution shape + box interior | sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted') |
| Violin Quartiles | sns.violinplot(inner='quart') | Quartiles inside violin | sns.violinplot(data=df, x='Gender', y='Salary', inner='quart', color='lightblue') |
| Split Violin | sns.violinplot(split=True) | Split by hue variable | sns.violinplot(data=df, x='Gender', y='Salary', hue='Smoker', split=True) |
| Box Plot | sns.boxplot() | Quartiles, median, outliers | sns.boxplot(data=df, x='Gender', y='Salary', width=0.5, palette='Set2') |
| Bar Plot (Mean) | sns.barplot() | Mean with confidence intervals | sns.barplot(data=df, x='Gender', y='Salary', ci='sd', capsize=0.1, palette='pastel') |
| Strip Plot | sns.stripplot() | Jittered individual points | sns.stripplot(data=df, x='Gender', y='Salary', jitter=True, alpha=0.6, color='black') |
| Swarm Plot | sns.swarmplot() | Non-overlapping points | sns.swarmplot(data=df, x='Gender', y='Salary', size=3, palette='dark') |
| Point Plot | sns.pointplot() | Means connected by lines | sns.pointplot(data=df, x='Gender', y='Salary', ci=95, color='red') |
| T-Test | scipy.stats.ttest_ind | Compare two group means | stats.ttest_ind(df[df.Gender=='M'].Salary, df[df.Gender=='F'].Salary) |
| ANOVA | scipy.stats.f_oneway | Compare 3+ group means | stats.f_oneway(*[g['Salary'].values for n,g in df.groupby('Department')]) |

INTERPRETATION - Categorical vs Continuous:
- Group means differ substantially: Category likely influences the continuous variable
- Violin shapes differ: Different distributions between groups
- Box plot medians far apart: Strong difference in central tendency
- Whiskers don't overlap: Groups likely statistically different
- CI bars don't overlap: Groups likely significantly different
- T-Test p < 0.05: Two group means are significantly different
- ANOVA p < 0.05: At least one group mean differs (post-hoc needed)

---

## CONTINUOUS VS CONTINUOUS

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Pearson Correlation | df.corr() | Linear correlation matrix | df[['Age','Salary','Experience']].corr() |
| Spearman Correlation | df.corr(method='spearman') | Rank-based monotonic correlation | df[['Age','Salary']].corr(method='spearman') |
| Kendall Correlation | df.corr(method='kendall') | Ordinal association | df[['Age','Salary']].corr(method='kendall') |
| NumPy Corrcoef | np.corrcoef() | Correlation coefficient matrix | np.corrcoef(df['Age'], df['Salary']) |
| Pearson + p-value | scipy.stats.pearsonr | Correlation with significance test | r, p = stats.pearsonr(df['Age'], df['Salary']) |
| Spearman + p-value | scipy.stats.spearmanr | Rank correlation with significance | rho, p = stats.spearmanr(df['Age'], df['Salary']) |
| Covariance | np.cov() | Joint variability matrix | np.cov(df['Age'], df['Salary'], ddof=1) |
| Pandas Covariance | df.cov() | Covariance matrix | df[['Age','Salary','Experience']].cov() |
| Scatter Plot | plt.scatter() | Basic scatter with size/color | plt.scatter(df['Age'], df['Salary'], s=100, c=df['Experience'], cmap='viridis', alpha=0.7, edgecolors='black') |
| Enhanced Scatter | sns.scatterplot() | Hue, size, style mapping | sns.scatterplot(data=df, x='Age', y='Salary', hue='Gender', size='Experience', style='Department') |
| Joint Scatter | sns.jointplot(kind='scatter') | Scatter + marginal distributions | sns.jointplot(data=df, x='Age', y='Salary', kind='scatter', hue='Gender') |
| Joint Regression | sns.jointplot(kind='reg') | Scatter + regression + KDE margins | sns.jointplot(data=df, x='Age', y='Salary', kind='reg', ci=95) |
| Hexbin Plot | sns.jointplot(kind='hex') | Density for large datasets | sns.jointplot(data=df, x='Age', y='Salary', kind='hex', gridsize=20) |
| 2D KDE | sns.jointplot(kind='kde') | Contour density plot | sns.jointplot(data=df, x='Age', y='Salary', kind='kde', fill=True) |
| Line Plot | plt.plot() | Basic line with styling | plt.plot(df['Year'], df['Revenue'], marker='o', linestyle='--', color='navy', linewidth=2) |
| Seaborn Line | sns.lineplot() | Line with confidence bands | sns.lineplot(data=df, x='Year', y='Revenue', hue='Region', ci=95, marker='o') |
| Pandas Line | df.plot() | Native line plot | df.plot(x='Year', y='Revenue', kind='line', figsize=(10,5), grid=True) |
| Correlation Heatmap | sns.heatmap() | Visual correlation matrix | sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True) |
| Masked Heatmap | sns.heatmap(mask=) | Upper triangle masked | mask = np.triu(np.ones_like(df.corr(), dtype=bool)); sns.heatmap(df.corr(), mask=mask, annot=True, cmap='RdBu_r') |
| Pair Plot | sns.pairplot() | All pairwise scatter + histograms | sns.pairplot(df[['Age','Salary','Experience']], diag_kind='kde', corner=True) |
| Pair Plot with Hue | sns.pairplot(hue=) | Color-coded by category | sns.pairplot(df, vars=['Age','Salary'], hue='Gender', palette='husl') |
| QQ Plot | scipy.stats.probplot() | Compare to theoretical distribution | stats.probplot(df['Salary'], dist="norm", plot=plt); plt.title('QQ Plot') |
| QQ Plot (Statsmodels) | sm.qqplot() | Alternative with reference line | sm.qqplot(df['Salary'], line='45', fit=True); plt.show() |
| Regression Plot | sns.regplot() | Scatter + regression line | sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5}, line_kws={'color':'red'}) |
| Faceted Regression | sns.lmplot() | Regression by category | sns.lmplot(data=df, x='Age', y='Salary', hue='Gender', col='Department', height=4) |
| Linear Regression | scipy.stats.linregress | Slope, intercept, r, p, std err | slope, intercept, r, p, se = stats.linregress(df['Age'], df['Salary']) |
| Describe | df.describe() | Count, mean, std, quartiles | df[['Age','Salary']].describe() |
| NumPy Mean/Std | np.mean(), np.std() | Central tendency and spread | np.mean(df['Salary']); np.std(df['Salary'], ddof=1) |
| Median/Percentile | np.median(), np.percentile() | Robust central tendency | np.median(df['Salary']); np.percentile(df['Salary'], [25, 50, 75]) |
| Quantile | np.quantile() | Quantile on 0-1 scale | np.quantile(df['Salary'], 0.95) |
| Full Stats | scipy.stats.describe() | nobs, minmax, mean, variance, skew, kurtosis | stats.describe(df['Salary']) |
| Skewness | scipy.stats.skew() | Asymmetry measure | stats.skew(df['Salary']) |
| Kurtosis | scipy.stats.kurtosis() | Tail heaviness | stats.kurtosis(df['Salary']) |
| Z-Score | scipy.stats.zscore() | Standardize to mean=0, std=1 | stats.zscore(df['Salary']) |

INTERPRETATION - Continuous vs Continuous:
- Pearson r ~ +1: Strong positive linear relationship
- Pearson r ~ -1: Strong negative linear relationship
- Pearson r ~ 0: No linear relationship (check non-linear!)
- Spearman rho ~ 1: Strong monotonic relationship
- Kendall tau ~ 1: Strong ordinal association
- p-value < 0.05: Correlation is statistically significant
- Heatmap ~ 1 (red): Strong positive correlation
- Heatmap ~ -1 (blue): Strong negative correlation
- Heatmap ~ 0 (neutral): No linear correlation
- QQ plot on 45 degree line: Data follows reference distribution
- QQ plot S-shape: Different tail behavior than reference
- QQ plot concave up: Right-skewed data
- QQ plot concave down: Left-skewed data
- Regression p < 0.05: Slope significantly different from zero
- R-squared value: Percentage of Y variance explained by X


## INSTALLATION

pip install pandas numpy matplotlib seaborn scipy statsmodels jupyter

---

## LICENSE

MIT License
```

```python
## COMPLETE WORKING EXAMPLE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

np.random.seed(42)
df = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], 200),
    'Department': np.random.choice(['Sales', 'IT', 'HR'], 200),
    'Age': np.random.normal(35, 10, 200).astype(int),
    'Salary': np.random.normal(70000, 15000, 200),
    'Experience': np.random.normal(10, 5, 200).astype(int)
})

# CATEGORICAL vs CATEGORICAL
ct = pd.crosstab(df['Gender'], df['Department'], margins=True, normalize='all')
chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['Gender'], df['Department']))
print(f"Chi-Square p-value: {p:.4f}")

# CATEGORICAL vs CONTINUOUS
print(df.groupby('Gender')['Salary'].agg(['mean', 'std', 'count']))
sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted')
plt.show()

# CONTINUOUS vs CONTINUOUS
corr = df[['Age', 'Salary', 'Experience']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.show()

sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5})
plt.show()

# QQ Plot
stats.probplot(df['Salary'], dist="norm", plot=plt)
plt.title('Q-Q Plot: Salary vs Normal Distribution')
plt.show()
```

```python
# =============================================================================
# ULTIMATE SCIPY.STATS CHEAT SHEET - Complete Reference
# =============================================================================
# A comprehensive reference for scipy.stats functions.
# Covers: Descriptive Statistics, Probability Distributions, Hypothesis Tests,
#         Normality Tests, Nonparametric Tests, Correlation & Regression,
#         Contingency Tables, and Resampling Methods.
#
# Format: Theme | Function | Explanation/Keywords | Syntax Example
# All examples are single-line, fully functional, and verified.
# =============================================================================

# -----------------------------------------------------------------------------
# IMPORTS
# -----------------------------------------------------------------------------
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =============================================================================
# SECTION 1: DESCRIPTIVE STATISTICS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Descriptive | stats.describe() | Full summary: nobs, minmax, mean, variance, skewness, kurtosis | stats.describe(data, ddof=1, bias=False) |
| Descriptive | stats.tmean() | Trimmed mean (excludes outliers from tails) | stats.tmean(data, limits=(0.1, 0.9)) |
| Descriptive | stats.tvar() | Trimmed variance | stats.tvar(data, limits=(0.1, 0.9)) |
| Descriptive | stats.tstd() | Trimmed standard deviation | stats.tstd(data, limits=(0.1, 0.9)) |
| Descriptive | stats.tsem() | Trimmed standard error of the mean | stats.tsem(data, limits=(0.1, 0.9)) |
| Descriptive | stats.gmean() | Geometric mean (for multiplicative/ratio data) | stats.gmean(data) |
| Descriptive | stats.hmean() | Harmonic mean (for rates/ratios) | stats.hmean(data) |
| Descriptive | stats.median_abs_deviation() | Robust MAD estimator (median absolute deviation) | stats.median_abs_deviation(data, scale='normal') |
| Descriptive | stats.iqr() | Interquartile range (75th - 25th percentile) | stats.iqr(data, rng=(25, 75)) |
| Descriptive | stats.sem() | Standard error of the mean | stats.sem(data, ddof=1) |
| Descriptive | stats.zscore() | Standardize to mean=0, std=1 | stats.zscore(data, ddof=1) |
| Descriptive | stats.skew() | Asymmetry measure (0=symmetric, Fisher) | stats.skew(data, bias=False) |
| Descriptive | stats.kurtosis() | Tail heaviness (0=normal, Fisher) | stats.kurtosis(data, fisher=True, bias=False) |
| Descriptive | stats.moment() | Non-central moment of order n | stats.moment(data, moment=3) |
| Descriptive | stats.variation() | Coefficient of variation (std/mean) | stats.variation(data, ddof=1) |
| Descriptive | stats.bootstrap() | Bootstrap confidence intervals for any statistic | stats.bootstrap((data,), np.mean, confidence_level=0.95, n_resamples=9999) |

---
INTERPRETATION - Descriptive Statistics:
- describe(): Compare sample stats to theoretical distribution params
- tmean/tvar: Use when outliers skew standard mean/variance
- gmean: Appropriate for growth rates, financial returns, ratios
- hmean: Appropriate for rates (speed, density)
- MAD: Robust alternative to std when data has heavy tails
- zscore: Values beyond |3| are typically considered outliers
- skew > 0: Right-skewed (long right tail); skew < 0: Left-skewed
- kurtosis > 0: Heavy tails (leptokurtic); < 0: Light tails (platykurtic)
- bootstrap CI: If interval excludes null value, effect is significant

# =============================================================================
# SECTION 2: PROBABILITY DISTRIBUTIONS (Continuous)
# =============================================================================
# All continuous distributions share these methods: pdf, cdf, sf, ppf, isf,
# rvs, mean, median, var, std, stats, moment, entropy, interval, fit, expect

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Normal | stats.norm | Gaussian distribution; loc=mean, scale=std | stats.norm.pdf(x, loc=0, scale=1) |
| Normal | stats.norm.cdf() | Cumulative probability up to x | stats.norm.cdf(1.96, loc=0, scale=1)  # returns ~0.975 |
| Normal | stats.norm.ppf() | Inverse CDF / quantile function | stats.norm.ppf(0.975, loc=0, scale=1)  # returns ~1.96 |
| Normal | stats.norm.rvs() | Random variates sampling | stats.norm.rvs(loc=0, scale=1, size=1000) |
| Normal | stats.norm.fit() | MLE parameter estimation from data | mu, sigma = stats.norm.fit(data) |
| Normal | stats.norm.interval() | Confidence interval for distribution | stats.norm.interval(0.95, loc=mu, scale=sigma) |
| t-dist | stats.t | Student's t-distribution; df=degrees of freedom | stats.t.pdf(x, df=10, loc=0, scale=1) |
| t-dist | stats.t.ppf() | Critical t-value for given alpha | stats.t.ppf(0.975, df=29)  # two-tailed 95% CI |
| Chi2 | stats.chi2 | Chi-squared distribution; df=degrees of freedom | stats.chi2.pdf(x, df=5) |
| Chi2 | stats.chi2.ppf() | Critical chi2 value | stats.chi2.ppf(0.95, df=5) |
| F-dist | stats.f | F-distribution; dfn, dfd = numerator/denominator df | stats.f.pdf(x, dfn=5, dfd=10) |
| F-dist | stats.f.ppf() | Critical F-value | stats.f.ppf(0.95, dfn=5, dfd=10) |
| Exponential | stats.expon | Exponential distribution; scale=1/lambda | stats.expon.pdf(x, scale=2) |
| Uniform | stats.uniform | Uniform distribution; loc=start, scale=width | stats.uniform.pdf(x, loc=0, scale=1) |
| Beta | stats.beta | Beta distribution; a, b = shape params | stats.beta.pdf(x, a=2, b=5) |
| Gamma | stats.gamma | Gamma distribution; a=shape, scale=1/rate | stats.gamma.pdf(x, a=2, scale=1) |
| Log-Normal | stats.lognorm | Log-normal; s=sigma, scale=exp(mu) | stats.lognorm.pdf(x, s=1, scale=np.exp(0)) |
| Weibull | stats.weibull_min | Weibull minimum; c=shape, scale=scale | stats.weibull_min.pdf(x, c=1.5, scale=1) |
| Cauchy | stats.cauchy | Cauchy/Lorentz distribution; heavy tails | stats.cauchy.pdf(x, loc=0, scale=1) |
| Laplace | stats.laplace | Double exponential distribution | stats.laplace.pdf(x, loc=0, scale=1) |
| Logistic | stats.logistic | Logistic distribution; S-shaped CDF | stats.logistic.pdf(x, loc=0, scale=1) |
| Pareto | stats.pareto | Pareto distribution; b=shape (power law) | stats.pareto.pdf(x, b=2.5) |

---
INTERPRETATION - Probability Distributions:
- pdf(x): Relative likelihood at point x (not probability!)
- cdf(x): P(X <= x), the cumulative probability
- sf(x): Survival function = 1 - cdf(x) = P(X > x)
- ppf(q): Quantile, returns x where cdf(x) = q
- isf(q): Inverse survival, returns x where sf(x) = q
- rvs(): Random samples; use size parameter for arrays
- fit(): Returns MLE estimates; compare AIC/BIC across distributions
- interval(alpha): Returns (lower, upper) for central probability alpha
- loc shifts distribution; scale stretches/compresses it

# =============================================================================
# SECTION 3: PROBABILITY DISTRIBUTIONS (Discrete)
# =============================================================================
# Discrete distributions use pmf instead of pdf; no scale parameter

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Binomial | stats.binom | n trials, p success probability | stats.binom.pmf(k=5, n=10, p=0.5) |
| Binomial | stats.binom.cdf() | Cumulative probability up to k successes | stats.binom.cdf(k=5, n=10, p=0.5) |
| Poisson | stats.poisson | lambda = rate parameter | stats.poisson.pmf(k=3, mu=2.5) |
| Geometric | stats.geom | p = success probability; number of trials until first success | stats.geom.pmf(k=3, p=0.2) |
| Negative Binomial | stats.nbinom | r successes, p probability; trials until r successes | stats.nbinom.pmf(k=5, n=3, p=0.5) |
| Hypergeometric | stats.hypergeom | M=population, n=successes, N=draws | stats.hypergeom.pmf(k=2, M=20, n=7, N=12) |
| Uniform (Discrete) | stats.randint | a=low, b=high+1 | stats.randint.pmf(k=3, low=1, high=6) |

---
INTERPRETATION - Discrete Distributions:
- pmf(k): Probability of exactly k occurrences
- cdf(k): Probability of k or fewer occurrences
- For binomial: use when independent trials with constant p
- For Poisson: use for rare events in fixed interval/space
- For geometric: use when counting trials until first success

# =============================================================================
# SECTION 4: NORMALITY TESTS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Shapiro-Wilk | stats.shapiro() | Most powerful for small samples (n < 50); tests normality | stat, p = stats.shapiro(data) |
| D'Agostino K2 | stats.normaltest() | Combines skewness and kurtosis; omnibus normality test | stat, p = stats.normaltest(data) |
| Jarque-Bera | stats.jarque_bera() | Tests skew=0 and kurtosis=3; needs n > 2000 | stat, p = stats.jarque_bera(data) |
| Kolmogorov-Smirnov | stats.kstest() | Compares sample to theoretical distribution | stat, p = stats.kstest(data, 'norm', args=(mu, sigma)) |
| Anderson-Darling | stats.anderson() | Sensitive to tail deviations; returns critical values | result = stats.anderson(data, dist='norm') |
| Cramer-von Mises | stats.cramervonmises() | Alternative to K-S; more powerful for some alternatives | stat, p = stats.cramervonmises(data, 'norm') |

---
INTERPRETATION - Normality Tests:
- Shapiro-Wilk: p > 0.05 => data likely normal (fail to reject H0)
- normaltest: p > 0.05 => data likely normal
- jarque_bera: p > 0.05 => data has normal skewness and kurtosis
- kstest: p > 0.05 => sample consistent with reference distribution
- anderson: Compare statistic to critical_values at chosen significance level
  - statistic < critical_value => fail to reject normality
- cramervonmises: p > 0.05 => fail to reject normality
- If ANY test rejects, consider data non-normal
- Small samples: prefer Shapiro-Wilk; large samples: prefer Anderson-Darling

# =============================================================================
# SECTION 5: PARAMETRIC HYPOTHESIS TESTS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| One-Sample t | stats.ttest_1samp() | Compare sample mean to known population mean | stat, p = stats.ttest_1samp(data, popmean=0) |
| Independent t | stats.ttest_ind() | Compare means of two independent groups | stat, p = stats.ttest_ind(group1, group2, equal_var=True) |
| Welch's t | stats.ttest_ind(equal_var=False) | Independent t without equal variance assumption | stat, p = stats.ttest_ind(group1, group2, equal_var=False) |
| Paired t | stats.ttest_rel() | Compare means of two paired/related samples | stat, p = stats.ttest_rel(before, after) |
| One-Way ANOVA | stats.f_oneway() | Compare means across 3+ independent groups | stat, p = stats.f_oneway(group1, group2, group3) |
| Levene | stats.levene() | Test equality of variances across groups | stat, p = stats.levene(group1, group2, group3) |
| Bartlett | stats.bartlett() | Test equality of variances (sensitive to non-normality) | stat, p = stats.bartlett(group1, group2, group3) |
| Fligner-Killeen | stats.fligner() | Non-parametric test for equal variances | stat, p = stats.fligner(group1, group2, group3) |

---
INTERPRETATION - Parametric Tests:
- ttest_1samp: p < 0.05 => sample mean differs from popmean
- ttest_ind: p < 0.05 => group means are significantly different
  - Use Welch's (equal_var=False) when variances differ
- ttest_rel: p < 0.05 => paired measurements differ significantly
- f_oneway: p < 0.05 => at least one group mean differs
  - Requires post-hoc tests (Tukey HSD) to identify which groups
- levene: p > 0.05 => equal variances assumption holds
- bartlett: p > 0.05 => equal variances (use only if normal)
- fligner: p > 0.05 => equal variances (robust to non-normality)

# =============================================================================
# SECTION 6: NONPARAMETRIC HYPOTHESIS TESTS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Mann-Whitney U | stats.mannwhitneyu() | Compare two independent samples (non-parametric t-test) | stat, p = stats.mannwhitneyu(group1, group2, alternative='two-sided') |
| Wilcoxon Signed-Rank | stats.wilcoxon() | Compare two paired/related samples (non-parametric paired t) | stat, p = stats.wilcoxon(before, after) |
| Kruskal-Wallis | stats.kruskal() | Compare 3+ independent groups (non-parametric ANOVA) | stat, p = stats.kruskal(group1, group2, group3) |
| Friedman | stats.friedmanchisquare() | Compare 3+ paired/related groups (non-parametric repeated measures) | stat, p = stats.friedmanchisquare(time1, time2, time3) |
| Sign Test | stats.wilcoxon(zero_method='zsplit') | Test median against hypothesized value | stat, p = stats.wilcoxon(data - hypothesized_median) |

---
INTERPRETATION - Nonparametric Tests:
- mannwhitneyu: p < 0.05 => distributions differ in location
  - alternative='less'/'greater' for one-tailed tests
- wilcoxon: p < 0.05 => paired samples differ in median
- kruskal: p < 0.05 => at least one group has different distribution
  - Follow with pairwise Mann-Whitney U with Bonferroni correction
- friedmanchisquare: p < 0.05 => at least one treatment differs
  - Use when same subjects measured under multiple conditions
- All rank-based: less power than parametric but more robust

# =============================================================================
# SECTION 7: CONTINGENCY TABLES & CATEGORICAL TESTS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Chi-Square | stats.chi2_contingency() | Test independence in contingency table | chi2, p, dof, expected = stats.chi2_contingency(observed_table) |
| Fisher Exact | stats.fisher_exact() | Exact test for 2x2 tables (small samples) | odds_ratio, p = stats.fisher_exact([[8, 2], [1, 5]]) |
| Barnard Exact | stats.barnard_exact() | More powerful alternative to Fisher for 2x2 | res = stats.barnard_exact([[8, 2], [1, 5]]) |
| Boschloo Exact | stats.boschloo_exact() | Most powerful exact test for 2x2 tables | res = stats.boschloo_exact([[8, 2], [1, 5]]) |
| Chi-Square Goodness | stats.chisquare() | Test if observed matches expected frequencies | stat, p = stats.chisquare(f_obs=observed, f_exp=expected) |
| Power Divergence | stats.power_divergence() | Generalized chi-square (Cressie-Read family) | stat, p = stats.power_divergence(f_obs, lambda_='log-likelihood') |
| Contingency Association | stats.contingency.association() | Measure association strength in contingency table | stats.contingency.association(observed, method='cramer') |
| Marginal Probabilities | stats.contingency.margins() | Compute row and column totals | row_totals, col_totals = stats.contingency.margins(observed) |

---
INTERPRETATION - Contingency Tests:
- chi2_contingency: p < 0.05 => variables are associated (not independent)
  - Expected frequencies should be >= 5 for validity
- fisher_exact: Use when expected cell counts < 5
  - Returns odds ratio: OR > 1 => positive association
- barnard_exact / boschloo_exact: More powerful than Fisher for 2x2
- chisquare: p < 0.05 => observed differs from expected distribution
- association(method='cramer'): Returns Cramer's V (0=none, 1=perfect)
- association(method='pearson'): Returns Pearson's contingency coefficient

# =============================================================================
# SECTION 8: CORRELATION & ASSOCIATION
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Pearson r | stats.pearsonr() | Linear correlation + p-value | r, p = stats.pearsonr(x, y) |
| Spearman rho | stats.spearmanr() | Rank-based monotonic correlation | rho, p = stats.spearmanr(x, y) |
| Kendall tau | stats.kendalltau() | Ordinal association (pairwise concordance) | tau, p = stats.kendalltau(x, y) |
| Point-Biserial | stats.pointbiserialr() | Correlation between binary and continuous | r, p = stats.pointbiserialr(binary_var, continuous_var) |
| Weighted Tau | stats.weightedtau() | Weighted Kendall's tau for ranked lists | tau, p = stats.weightedtau(x, y, rank=True) |
| Somers' D | stats.somersd() | Asymmetric association measure | res = stats.somersd(x, y) |

---
INTERPRETATION - Correlation:
- pearsonr: r ~ +1 strong positive linear; r ~ -1 strong negative; r ~ 0 no linear
  - p < 0.05 => correlation is statistically significant
- spearmanr: rho ~ +1 strong monotonic increasing; rho ~ -1 strong monotonic decreasing
  - Use when relationship is monotonic but not linear
- kendalltau: tau ~ +1 perfect agreement; tau ~ -1 perfect disagreement
  - More robust than Spearman for small samples and ties
- pointbiserialr: Use when one variable is dichotomous (0/1)
- weightedtau: Use for comparing ranked lists with different importance weights
- somersd: Asymmetric; useful when one variable predicts another

# =============================================================================
# SECTION 9: LINEAR REGRESSION & MODELING
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Linear Regression | stats.linregress() | Simple linear regression: slope, intercept, r, p, stderr | slope, intercept, r, p, se = stats.linregress(x, y) |
| Theil-Sen | stats.theilslopes() | Robust regression (median of all pairwise slopes) | res = stats.theilslopes(y, x, confidence=0.95) |
| Siegel | stats.siegelslopes() | Robust regression (repeated median) | slope, intercept = stats.siegelslopes(y, x) |
| Sensitivity Analysis | stats.sensitivity_stats() | Slope sensitivity analysis | stats.sensitivity_stats(x, y) |

---
INTERPRETATION - Regression:
- linregress: p < 0.05 => slope significantly different from zero
  - r_value**2 = R-squared (proportion of variance explained)
  - stderr = standard error of slope estimate
- theilslopes: More robust to outliers than OLS; use when outliers present
  - Returns confidence interval for slope
- siegelslopes: Even more robust than Theil-Sen for heavily contaminated data

# =============================================================================
# SECTION 10: RESAMPLING & PERMUTATION METHODS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Bootstrap CI | stats.bootstrap() | Non-parametric confidence intervals | res = stats.bootstrap((data,), np.median, confidence_level=0.95, n_resamples=9999) |
| Permutation Test | stats.permutation_test() | Exact/randomized permutation p-values | def statistic(x, y): return np.mean(x) - np.mean(y); res = stats.permutation_test((group1, group2), statistic, n_resamples=10000) |
| Monte Carlo | stats.monte_carlo_test() | Monte Carlo hypothesis testing | res = stats.monte_carlo_test(data, stats.norm.rvs, lambda x: np.mean(x), n_resamples=10000) |

---
INTERPRETATION - Resampling:
- bootstrap: CI that excludes null => significant effect
  - More accurate than parametric methods for complex statistics
- permutation_test: p < 0.05 => groups differ significantly
  - Makes no distributional assumptions; exact when enumerating all permutations
- monte_carlo_test: p < 0.05 => data inconsistent with null model

# =============================================================================
# SECTION 11: MULTIPLE COMPARISONS & POST-HOC
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Bonferroni | stats.false_discovery_rate() | Control family-wise error rate | No direct function; divide alpha by number of tests |
| False Discovery Rate | No direct scipy function | Control FDR (use statsmodels) | from statsmodels.stats.multitest import multipletests; reject, p_adj, _, _ = multipletests(p_values, method='fdr_bh') |
| Tukey HSD | No direct scipy function | Post-hoc for ANOVA (use statsmodels) | from statsmodels.stats.multicomp import pairwise_tukeyhsd; tukey = pairwise_tukeyhsd(data, groups, alpha=0.05) |

---
INTERPRETATION - Multiple Comparisons:
- Bonferroni: Conservative; multiply p-values by number of tests
- FDR (Benjamini-Hochberg): Less conservative; controls false discovery rate
- Tukey HSD: Compare all pairs while controlling family-wise error rate
  - Reject null for pairs where "reject" column is True

# =============================================================================
# SECTION 12: SPECIALIZED STATISTICS
# =============================================================================

| Theme | Function | Explanation/Keywords | Syntax Example |
|-------|----------|----------------------|----------------|
| Trimmed Stats | stats.trimboth() | Trim symmetric proportion from both ends | trimmed = stats.trimboth(data, proportiontocut=0.1) |
| Winsorize | stats.mstats.winsorize() | Cap extreme values instead of removing | winsorized = stats.mstats.winsorize(data, limits=[0.05, 0.05]) |
| Rank Data | stats.rankdata() | Assign ranks to data (handle ties) | ranks = stats.rankdata(data, method='average') |
| Percentile | stats.scoreatpercentile() | Value at given percentile | val = stats.scoreatpercentile(data, 75) |
| Percentile | stats.percentileofscore() | Percentile rank of a score | pct = stats.percentileofscore(data, score=50) |
| Mode | stats.mode() | Most frequent value(s) | mode_res = stats.mode(data, keepdims=True) |
| Entropy | stats.entropy() | Shannon entropy / Kullback-Leibler divergence | H = stats.entropy(pk, base=2) |
| Rel Entropy | stats.entropy() | KL divergence between two distributions | KL = stats.entropy(pk, qk) |
| Energy Distance | stats.energy_distance() | Distance between two distributions | dist = stats.energy_distance(data1, data2) |
| Wasserstein | stats.wasserstein_distance() | Earth mover's distance | dist = stats.wasserstein_distance(data1, data2) |

---
INTERPRETATION - Specialized Statistics:
- trimboth: Removes outliers symmetrically; use for robust analysis
- winsorize: Caps outliers; retains sample size unlike trimming
- rankdata: Essential for nonparametric tests; method='average' for ties
- entropy: Higher values = more uncertainty/disorder
- energy_distance: Zero => identical distributions; increases with difference
- wasserstein_distance: Interpretable as minimum "work" to transform one dist to another

# =============================================================================
# COMPLETE WORKING EXAMPLE
# =============================================================================

from scipy import stats
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate sample data
group_a = np.random.normal(100, 15, 50)
group_b = np.random.normal(110, 15, 50)
before = np.random.normal(100, 15, 50)
after = before + np.random.normal(5, 5, 50)

# DESCRIPTIVE
print(stats.describe(group_a))

# NORMALITY
shapiro_stat, shapiro_p = stats.shapiro(group_a)
print(f"Shapiro-Wilk p: {shapiro_p:.4f}")

# PARAMETRIC TEST
ind_t_stat, ind_t_p = stats.ttest_ind(group_a, group_b)
print(f"Independent t-test p: {ind_t_p:.4f}")

# NONPARAMETRIC TEST
mw_stat, mw_p = stats.mannwhitneyu(group_a, group_b)
print(f"Mann-Whitney U p: {mw_p:.4f}")

# PAIRED TEST
paired_t_stat, paired_t_p = stats.ttest_rel(before, after)
print(f"Paired t-test p: {paired_t_p:.4f}")

# CORRELATION
r, p = stats.pearsonr(group_a, group_b)
print(f"Pearson r: {r:.4f}, p: {p:.4f}")

# REGRESSION
slope, intercept, r_val, p_val, std_err = stats.linregress(group_a, group_b)
print(f"Slope: {slope:.4f}, R-squared: {r_val**2:.4f}")

# BOOTSTRAP CI
res = stats.bootstrap((group_a,), np.mean, confidence_level=0.95, n_resamples=9999)
print(f"95% CI for mean: ({res.confidence_interval.low:.2f}, {res.confidence_interval.high:.2f})")

# =============================================================================
# INSTALLATION
# =============================================================================
# pip install scipy numpy pandas matplotlib
# =============================================================================
```

```python
## Quick Reference: Common Patterns

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Sample data
df = pd.DataFrame({
    'Gender': np.random.choice(['M','F'], 200),
    'Department': np.random.choice(['Sales','IT','HR'], 200),
    'Age': np.random.normal(35, 10, 200).astype(int),
    'Salary': np.random.normal(70000, 15000, 200),
    'Experience': np.random.normal(10, 5, 200).astype(int)
})

# === CATEGORICAL VS CATEGORICAL ===
# Crosstab with margins
ct = pd.crosstab(df['Gender'], df['Department'], margins=True, normalize='all')

# Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['Gender'], df['Department']))

# Cramer's V
cont = pd.crosstab(df['Gender'], df['Department'])
n = cont.sum().sum()
phi2 = stats.chi2_contingency(cont)[0] / n
r, k = cont.shape
phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
rcorr = r - ((r-1)**2)/(n-1)
kcorr = k - ((k-1)**2)/(n-1)
cramers_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1))) if min((kcorr-1), (rcorr-1)) > 0 else 0

# Stacked bar
ct.iloc[:-1, :-1].plot(kind='bar', stacked=True, colormap='Set2')

# Frequency heatmap
sns.heatmap(ct.iloc[:-1, :-1], annot=True, fmt='.2f', cmap='YlOrRd')

# === CATEGORICAL VS CONTINUOUS ===
# Group statistics
grouped = df.groupby('Gender')['Salary'].agg(['mean','std','count','min','max'])

# Violin plot
sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted')

# Box plot
sns.boxplot(data=df, x='Department', y='Salary', palette='Set3')

# Bar plot with CI
sns.barplot(data=df, x='Gender', y='Salary', ci='sd', capsize=0.1)

# T-test
t_stat, t_p = stats.ttest_ind(df[df.Gender=='M'].Salary, df[df.Gender=='F'].Salary)

# === CONTINUOUS VS CONTINUOUS ===
# Correlation matrix
corr = df[['Age','Salary','Experience']].corr()

# Correlation heatmap
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True)

# Scatter plot
sns.scatterplot(data=df, x='Age', y='Salary', hue='Gender', size='Experience', alpha=0.7)

# Regression plot
sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5})

# Joint plot
sns.jointplot(data=df, x='Age', y='Salary', kind='reg', ci=95)

# QQ plot
stats.probplot(df['Salary'], dist="norm", plot=plt)
plt.title('Q-Q Plot: Salary vs Normal Distribution')

# Pair plot
sns.pairplot(df[['Age','Salary','Experience']], diag_kind='kde', corner=True)

# Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Age'], df['Salary'])
```

```python
##How to Interpret the Results

Categorical vs Categorical
| Result                        | Interpretation                                                                                      |
| ----------------------------- | --------------------------------------------------------------------------------------------------- |
| **Crosstab counts**           | Higher counts in specific cells indicate association between those categories                       |
| **Normalized crosstab**       | Proportions reveal relative frequencies; row-normalize to see distribution within each row category |
| **Chi-Square p-value < 0.05** | Reject independence — the two categorical variables are **statistically associated**                |
| **Chi-Square p-value ≥ 0.05** | Fail to reject independence — no significant association detected                                   |
| **Cramér's V ≈ 0**            | No association between variables                                                                    |
| **Cramér's V ≈ 0.1**          | Weak association                                                                                    |
| **Cramér's V ≈ 0.3**          | Moderate association                                                                                |
| **Cramér's V ≈ 0.5+**         | Strong association                                                                                  |
| **Stacked/Grouped Bars**      | Visual comparison of category frequencies; unequal bar heights suggest association                  |
| **Frequency Heatmap**         | Darker/brighter cells indicate higher frequencies; patterns reveal relationships                    |

Categorical vs Continuous
| Result                                          | Interpretation                                                              |
| ----------------------------------------------- | --------------------------------------------------------------------------- |
| **Group means differ substantially**            | The categorical variable likely influences the continuous variable          |
| **Violin plot shapes differ**                   | Different distributions (e.g., one group bimodal, another skewed)           |
| **Box plot medians far apart**                  | Strong difference in central tendency between groups                        |
| **Box plot whiskers don't overlap**             | Groups are likely statistically different                                   |
| **Box plot outliers**                           | Points beyond whiskers (1.5×IQR) may be anomalies or genuine extreme values |
| **Bar plot confidence intervals overlap**       | Groups may NOT be significantly different                                   |
| **Bar plot confidence intervals don't overlap** | Groups are likely significantly different                                   |
| **T-Test p-value < 0.05**                       | Two group means are significantly different                                 |
| **ANOVA p-value < 0.05**                        | At least one group mean differs from others (post-hoc tests needed)         |


Continuous vs Continuous
| Result                            | Interpretation                                            |
| --------------------------------- | --------------------------------------------------------- |
| **Pearson r ≈ +1**                | Strong positive linear relationship                       |
| **Pearson r ≈ -1**                | Strong negative linear relationship                       |
| **Pearson r ≈ 0**                 | No linear relationship (but may have non-linear!)         |
| **Spearman ρ ≈ 1**                | Strong monotonic relationship (rank-based)                |
| **Kendall τ ≈ 1**                 | Strong ordinal association                                |
| **Correlation p-value < 0.05**    | Correlation is statistically significant                  |
| **Scatter points form a line**    | Strong linear relationship                                |
| **Scatter points form a curve**   | Non-linear relationship (consider transformations)        |
| **Scatter points form a cloud**   | Weak or no relationship                                   |
| **Heatmap cell ≈ 1 (dark red)**   | Strong positive correlation                               |
| **Heatmap cell ≈ -1 (dark blue)** | Strong negative correlation                               |
| **Heatmap cell ≈ 0 (neutral)**    | No linear correlation                                     |
| **QQ plot points on 45° line**    | Data follows the theoretical distribution (e.g., normal)  |
| **QQ plot S-shape**               | Heavier or lighter tails than reference distribution      |
| **QQ plot concave up (smile)**    | Right-skewed data                                         |
| **QQ plot concave down (frown)**  | Left-skewed data                                          |
| **Regression p-value < 0.05**     | Slope is significantly different from zero                |
| **R² value**                      | Proportion of variance in Y explained by X (0-1 scale)    |
```

```python
#Drop Outliers


#drop all bigger than 170
df = df[df["column"] > 170]

#Keep rows at or below 170
df = df[df["column"] <= 170]

#Drop rows above 170
df = df.drop(df[df["column"] > 170].index)

#If you meant one column name
df = df[df["column"] <= 170]

```

```python

# happens because  p_value  is not a number, it’s a  TtestResult  object, so you cannot compare it directly with  alpha = 0.05 .
#How to fix your code. . . stats.ttest_ind(...)  returns a named tuple  TtestResult(statistic, pvalue, ...) , so you must extract the  pvalue :

import scipy.stats as stats

alpha = 0.05

# Do the test
result = stats.ttest_ind(df[df.type_1=='Dragon'].hp, df[df.type_1=='Grass'].hp)

# Extract the p-value
p_value = result.pvalue

if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypothesis")


# --> Alternatively in one line version

p_value = stats.ttest_ind(df[df.type_1=='Dragon'].hp, df[df.type_1=='Grass'].hp).pvalue

if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypothesis")
```

```python
#code here
H0: lengendar stats = legendary_not stats
H0: lengendar stats >= legendary_not stats

legend = df[df["legendary"]==True]
#hp  attack  defense  sp_atk  sp_def  speed

legend_not = df[df["legendary"]==False]
#hp  attack  defense  sp_atk  sp_def  speed

p_hp = stats.ttest_ind(legend["hp"], legend_not["hp"], equal_var=False).pvalue
p_attack = stats.ttest_ind(legend["attack"], legend_not["attack"], equal_var=False).pvalue
p_defense = stats.ttest_ind(legend["defense"], legend_not["defense"], equal_var=False).pvalue
p_sp_atk = stats.ttest_ind(legend["sp_atk"], legend_not["sp_atk"], equal_var=False).pvalue
p_sp_def = stats.ttest_ind(legend["sp_def"], legend_not["sp_def"], equal_var=False).pvalue
p_speed = stats.ttest_ind(legend["speed"], legend_not["speed"], equal_var=False).pvalue


if p_hp > alpha:
    print("We are not able to reject the null hypothesis")
if p_attack > alpha:
    print("We are not able to reject the null hypothesis")
if p_defense > alpha:
    print("We are not able to reject the null hypothesis")
if p_sp_atk > alpha:
    print("We are not able to reject the null hypothesis")
if p_sp_def > alpha:
    print("We are not able to reject the null hypothesis")
if p_speed > alpha:
    print("We are not able to reject the null hypothesis")

if p_hp < alpha:
    print("ph_We are not able to reject the null hypothesis")
if p_attack < alpha:
    print("p_attack We are  able to reject the null hypothesis")
if p_defense < alpha:
    print("p_defense We are  able to reject the null hypothesis")
if p_sp_atk < alpha:
    print(" p_sp_atk We  not able to reject the null hypothesis")
if p_sp_def < alpha:
    print("p_sp_def We  not able to reject the null hypothesis")
if p_speed < alpha:
    print("p_speed We  not able to reject the null hypothesis")
else:
    print("nothing")
```

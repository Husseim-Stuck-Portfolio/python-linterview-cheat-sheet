```python
# Ultimate Python Statistics Cheat Sheet

#A comprehensive reference for data analysis with Pandas, NumPy, Matplotlib, and Seaborn. 
#Covers Categorical vs Categorical, Categorical vs Continuous, and Continuous vs Continuous relationships.

---

## IMPORTS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.graphics.mosaicplot import mosaic
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

---

## CATEGORICAL VS CATEGORICAL

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Crosstab | pd.crosstab() | Frequency table with margins and normalization | pd.crosstab(df['Gender'], df['Product'], margins=True, normalize='all') |
| Crosstab Row Norm | pd.crosstab(normalize='index') | Proportions within each row category | pd.crosstab(df['Gender'], df['Product'], normalize='index') |
| Crosstab Col Norm | pd.crosstab(normalize='columns') | Proportions within each column category | pd.crosstab(df['Gender'], df['Product'], normalize='columns') |
| Crosstab Aggregate | pd.crosstab(aggfunc=) | Aggregate values by categories | pd.crosstab(df['Gender'], df['Product'], values=df['Sales'], aggfunc='mean') |
| Chi-Square Test | scipy.stats.chi2_contingency | Test independence between two categorical variables | chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['A'], df['B'])) |
| Cramer's V | Manual calculation | Association strength (0=none, 1=perfect) | n = cont.sum().sum(); phi2 = chi2/n; r, k = cont.shape; phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1)); rcorr = r-((r-1)**2)/(n-1); kcorr = k-((k-1)**2)/(n-1); v = np.sqrt(phi2corr/min((kcorr-1),(rcorr-1))) |
| Stacked Bar | df.plot(kind='bar', stacked=True) | Composition of categories | pd.crosstab(df['Gender'], df['Product']).plot(kind='bar', stacked=True, colormap='Set2') |
| Grouped Bar | df.plot(kind='bar') | Side-by-side comparison | pd.crosstab(df['Gender'], df['Product']).plot(kind='bar', color=['skyblue','salmon']) |
| Frequency Heatmap | sns.heatmap() | Color-coded frequency matrix | sns.heatmap(pd.crosstab(df['Gender'], df['Product']), annot=True, fmt='d', cmap='YlOrRd') |
| Mosaic Plot | statsmodels.graphics.mosaicplot | Proportional tile visualization | from statsmodels.graphics.mosaicplot import mosaic; mosaic(df, ['Gender','Product']) |

INTERPRETATION - Categorical vs Categorical:
- Chi-Square p < 0.05: Variables are statistically associated (reject independence)
- Chi-Square p >= 0.05: No significant association detected
- Cramer's V ~ 0: No association
- Cramer's V ~ 0.1: Weak association
- Cramer's V ~ 0.3: Moderate association
- Cramer's V ~ 0.5+: Strong association
- Unequal bar heights: Visual evidence of association between categories

---

## CATEGORICAL VS CONTINUOUS

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Group Stats | df.groupby().agg() | Aggregate by category | df.groupby('Gender')['Salary'].agg(['mean','std','count','min','max']) |
| Group Describe | df.groupby().describe() | Full descriptive stats per group | df.groupby('Gender')['Salary'].describe() |
| Violin Plot | sns.violinplot() | Distribution shape + box interior | sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted') |
| Violin Quartiles | sns.violinplot(inner='quart') | Quartiles inside violin | sns.violinplot(data=df, x='Gender', y='Salary', inner='quart', color='lightblue') |
| Split Violin | sns.violinplot(split=True) | Split by hue variable | sns.violinplot(data=df, x='Gender', y='Salary', hue='Smoker', split=True) |
| Box Plot | sns.boxplot() | Quartiles, median, outliers | sns.boxplot(data=df, x='Gender', y='Salary', width=0.5, palette='Set2') |
| Bar Plot (Mean) | sns.barplot() | Mean with confidence intervals | sns.barplot(data=df, x='Gender', y='Salary', ci='sd', capsize=0.1, palette='pastel') |
| Strip Plot | sns.stripplot() | Jittered individual points | sns.stripplot(data=df, x='Gender', y='Salary', jitter=True, alpha=0.6, color='black') |
| Swarm Plot | sns.swarmplot() | Non-overlapping points | sns.swarmplot(data=df, x='Gender', y='Salary', size=3, palette='dark') |
| Point Plot | sns.pointplot() | Means connected by lines | sns.pointplot(data=df, x='Gender', y='Salary', ci=95, color='red') |
| T-Test | scipy.stats.ttest_ind | Compare two group means | stats.ttest_ind(df[df.Gender=='M'].Salary, df[df.Gender=='F'].Salary) |
| ANOVA | scipy.stats.f_oneway | Compare 3+ group means | stats.f_oneway(*[g['Salary'].values for n,g in df.groupby('Department')]) |

INTERPRETATION - Categorical vs Continuous:
- Group means differ substantially: Category likely influences the continuous variable
- Violin shapes differ: Different distributions between groups
- Box plot medians far apart: Strong difference in central tendency
- Whiskers don't overlap: Groups likely statistically different
- CI bars don't overlap: Groups likely significantly different
- T-Test p < 0.05: Two group means are significantly different
- ANOVA p < 0.05: At least one group mean differs (post-hoc needed)

---

## CONTINUOUS VS CONTINUOUS

| Theme | Function | Explanation | Syntax Example |
|-------|----------|-------------|----------------|
| Pearson Correlation | df.corr() | Linear correlation matrix | df[['Age','Salary','Experience']].corr() |
| Spearman Correlation | df.corr(method='spearman') | Rank-based monotonic correlation | df[['Age','Salary']].corr(method='spearman') |
| Kendall Correlation | df.corr(method='kendall') | Ordinal association | df[['Age','Salary']].corr(method='kendall') |
| NumPy Corrcoef | np.corrcoef() | Correlation coefficient matrix | np.corrcoef(df['Age'], df['Salary']) |
| Pearson + p-value | scipy.stats.pearsonr | Correlation with significance test | r, p = stats.pearsonr(df['Age'], df['Salary']) |
| Spearman + p-value | scipy.stats.spearmanr | Rank correlation with significance | rho, p = stats.spearmanr(df['Age'], df['Salary']) |
| Covariance | np.cov() | Joint variability matrix | np.cov(df['Age'], df['Salary'], ddof=1) |
| Pandas Covariance | df.cov() | Covariance matrix | df[['Age','Salary','Experience']].cov() |
| Scatter Plot | plt.scatter() | Basic scatter with size/color | plt.scatter(df['Age'], df['Salary'], s=100, c=df['Experience'], cmap='viridis', alpha=0.7, edgecolors='black') |
| Enhanced Scatter | sns.scatterplot() | Hue, size, style mapping | sns.scatterplot(data=df, x='Age', y='Salary', hue='Gender', size='Experience', style='Department') |
| Joint Scatter | sns.jointplot(kind='scatter') | Scatter + marginal distributions | sns.jointplot(data=df, x='Age', y='Salary', kind='scatter', hue='Gender') |
| Joint Regression | sns.jointplot(kind='reg') | Scatter + regression + KDE margins | sns.jointplot(data=df, x='Age', y='Salary', kind='reg', ci=95) |
| Hexbin Plot | sns.jointplot(kind='hex') | Density for large datasets | sns.jointplot(data=df, x='Age', y='Salary', kind='hex', gridsize=20) |
| 2D KDE | sns.jointplot(kind='kde') | Contour density plot | sns.jointplot(data=df, x='Age', y='Salary', kind='kde', fill=True) |
| Line Plot | plt.plot() | Basic line with styling | plt.plot(df['Year'], df['Revenue'], marker='o', linestyle='--', color='navy', linewidth=2) |
| Seaborn Line | sns.lineplot() | Line with confidence bands | sns.lineplot(data=df, x='Year', y='Revenue', hue='Region', ci=95, marker='o') |
| Pandas Line | df.plot() | Native line plot | df.plot(x='Year', y='Revenue', kind='line', figsize=(10,5), grid=True) |
| Correlation Heatmap | sns.heatmap() | Visual correlation matrix | sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True) |
| Masked Heatmap | sns.heatmap(mask=) | Upper triangle masked | mask = np.triu(np.ones_like(df.corr(), dtype=bool)); sns.heatmap(df.corr(), mask=mask, annot=True, cmap='RdBu_r') |
| Pair Plot | sns.pairplot() | All pairwise scatter + histograms | sns.pairplot(df[['Age','Salary','Experience']], diag_kind='kde', corner=True) |
| Pair Plot with Hue | sns.pairplot(hue=) | Color-coded by category | sns.pairplot(df, vars=['Age','Salary'], hue='Gender', palette='husl') |
| QQ Plot | scipy.stats.probplot() | Compare to theoretical distribution | stats.probplot(df['Salary'], dist="norm", plot=plt); plt.title('QQ Plot') |
| QQ Plot (Statsmodels) | sm.qqplot() | Alternative with reference line | sm.qqplot(df['Salary'], line='45', fit=True); plt.show() |
| Regression Plot | sns.regplot() | Scatter + regression line | sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5}, line_kws={'color':'red'}) |
| Faceted Regression | sns.lmplot() | Regression by category | sns.lmplot(data=df, x='Age', y='Salary', hue='Gender', col='Department', height=4) |
| Linear Regression | scipy.stats.linregress | Slope, intercept, r, p, std err | slope, intercept, r, p, se = stats.linregress(df['Age'], df['Salary']) |
| Describe | df.describe() | Count, mean, std, quartiles | df[['Age','Salary']].describe() |
| NumPy Mean/Std | np.mean(), np.std() | Central tendency and spread | np.mean(df['Salary']); np.std(df['Salary'], ddof=1) |
| Median/Percentile | np.median(), np.percentile() | Robust central tendency | np.median(df['Salary']); np.percentile(df['Salary'], [25, 50, 75]) |
| Quantile | np.quantile() | Quantile on 0-1 scale | np.quantile(df['Salary'], 0.95) |
| Full Stats | scipy.stats.describe() | nobs, minmax, mean, variance, skew, kurtosis | stats.describe(df['Salary']) |
| Skewness | scipy.stats.skew() | Asymmetry measure | stats.skew(df['Salary']) |
| Kurtosis | scipy.stats.kurtosis() | Tail heaviness | stats.kurtosis(df['Salary']) |
| Z-Score | scipy.stats.zscore() | Standardize to mean=0, std=1 | stats.zscore(df['Salary']) |

INTERPRETATION - Continuous vs Continuous:
- Pearson r ~ +1: Strong positive linear relationship
- Pearson r ~ -1: Strong negative linear relationship
- Pearson r ~ 0: No linear relationship (check non-linear!)
- Spearman rho ~ 1: Strong monotonic relationship
- Kendall tau ~ 1: Strong ordinal association
- p-value < 0.05: Correlation is statistically significant
- Heatmap ~ 1 (red): Strong positive correlation
- Heatmap ~ -1 (blue): Strong negative correlation
- Heatmap ~ 0 (neutral): No linear correlation
- QQ plot on 45 degree line: Data follows reference distribution
- QQ plot S-shape: Different tail behavior than reference
- QQ plot concave up: Right-skewed data
- QQ plot concave down: Left-skewed data
- Regression p < 0.05: Slope significantly different from zero
- R-squared value: Percentage of Y variance explained by X


## INSTALLATION

pip install pandas numpy matplotlib seaborn scipy statsmodels jupyter

---

## LICENSE

MIT License
```

```python
## COMPLETE WORKING EXAMPLE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

np.random.seed(42)
df = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], 200),
    'Department': np.random.choice(['Sales', 'IT', 'HR'], 200),
    'Age': np.random.normal(35, 10, 200).astype(int),
    'Salary': np.random.normal(70000, 15000, 200),
    'Experience': np.random.normal(10, 5, 200).astype(int)
})

# CATEGORICAL vs CATEGORICAL
ct = pd.crosstab(df['Gender'], df['Department'], margins=True, normalize='all')
chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['Gender'], df['Department']))
print(f"Chi-Square p-value: {p:.4f}")

# CATEGORICAL vs CONTINUOUS
print(df.groupby('Gender')['Salary'].agg(['mean', 'std', 'count']))
sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted')
plt.show()

# CONTINUOUS vs CONTINUOUS
corr = df[['Age', 'Salary', 'Experience']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.show()

sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5})
plt.show()

# QQ Plot
stats.probplot(df['Salary'], dist="norm", plot=plt)
plt.title('Q-Q Plot: Salary vs Normal Distribution')
plt.show()
```

```python
## Quick Reference: Common Patterns

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Sample data
df = pd.DataFrame({
    'Gender': np.random.choice(['M','F'], 200),
    'Department': np.random.choice(['Sales','IT','HR'], 200),
    'Age': np.random.normal(35, 10, 200).astype(int),
    'Salary': np.random.normal(70000, 15000, 200),
    'Experience': np.random.normal(10, 5, 200).astype(int)
})

# === CATEGORICAL VS CATEGORICAL ===
# Crosstab with margins
ct = pd.crosstab(df['Gender'], df['Department'], margins=True, normalize='all')

# Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df['Gender'], df['Department']))

# Cramer's V
cont = pd.crosstab(df['Gender'], df['Department'])
n = cont.sum().sum()
phi2 = stats.chi2_contingency(cont)[0] / n
r, k = cont.shape
phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
rcorr = r - ((r-1)**2)/(n-1)
kcorr = k - ((k-1)**2)/(n-1)
cramers_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1))) if min((kcorr-1), (rcorr-1)) > 0 else 0

# Stacked bar
ct.iloc[:-1, :-1].plot(kind='bar', stacked=True, colormap='Set2')

# Frequency heatmap
sns.heatmap(ct.iloc[:-1, :-1], annot=True, fmt='.2f', cmap='YlOrRd')

# === CATEGORICAL VS CONTINUOUS ===
# Group statistics
grouped = df.groupby('Gender')['Salary'].agg(['mean','std','count','min','max'])

# Violin plot
sns.violinplot(data=df, x='Gender', y='Salary', inner='box', palette='muted')

# Box plot
sns.boxplot(data=df, x='Department', y='Salary', palette='Set3')

# Bar plot with CI
sns.barplot(data=df, x='Gender', y='Salary', ci='sd', capsize=0.1)

# T-test
t_stat, t_p = stats.ttest_ind(df[df.Gender=='M'].Salary, df[df.Gender=='F'].Salary)

# === CONTINUOUS VS CONTINUOUS ===
# Correlation matrix
corr = df[['Age','Salary','Experience']].corr()

# Correlation heatmap
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, square=True)

# Scatter plot
sns.scatterplot(data=df, x='Age', y='Salary', hue='Gender', size='Experience', alpha=0.7)

# Regression plot
sns.regplot(data=df, x='Age', y='Salary', ci=95, scatter_kws={'alpha':0.5})

# Joint plot
sns.jointplot(data=df, x='Age', y='Salary', kind='reg', ci=95)

# QQ plot
stats.probplot(df['Salary'], dist="norm", plot=plt)
plt.title('Q-Q Plot: Salary vs Normal Distribution')

# Pair plot
sns.pairplot(df[['Age','Salary','Experience']], diag_kind='kde', corner=True)

# Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Age'], df['Salary'])
```

```python
##How to Interpret the Results

Categorical vs Categorical
| Result                        | Interpretation                                                                                      |
| ----------------------------- | --------------------------------------------------------------------------------------------------- |
| **Crosstab counts**           | Higher counts in specific cells indicate association between those categories                       |
| **Normalized crosstab**       | Proportions reveal relative frequencies; row-normalize to see distribution within each row category |
| **Chi-Square p-value < 0.05** | Reject independence — the two categorical variables are **statistically associated**                |
| **Chi-Square p-value ≥ 0.05** | Fail to reject independence — no significant association detected                                   |
| **Cramér's V ≈ 0**            | No association between variables                                                                    |
| **Cramér's V ≈ 0.1**          | Weak association                                                                                    |
| **Cramér's V ≈ 0.3**          | Moderate association                                                                                |
| **Cramér's V ≈ 0.5+**         | Strong association                                                                                  |
| **Stacked/Grouped Bars**      | Visual comparison of category frequencies; unequal bar heights suggest association                  |
| **Frequency Heatmap**         | Darker/brighter cells indicate higher frequencies; patterns reveal relationships                    |

Categorical vs Continuous
| Result                                          | Interpretation                                                              |
| ----------------------------------------------- | --------------------------------------------------------------------------- |
| **Group means differ substantially**            | The categorical variable likely influences the continuous variable          |
| **Violin plot shapes differ**                   | Different distributions (e.g., one group bimodal, another skewed)           |
| **Box plot medians far apart**                  | Strong difference in central tendency between groups                        |
| **Box plot whiskers don't overlap**             | Groups are likely statistically different                                   |
| **Box plot outliers**                           | Points beyond whiskers (1.5×IQR) may be anomalies or genuine extreme values |
| **Bar plot confidence intervals overlap**       | Groups may NOT be significantly different                                   |
| **Bar plot confidence intervals don't overlap** | Groups are likely significantly different                                   |
| **T-Test p-value < 0.05**                       | Two group means are significantly different                                 |
| **ANOVA p-value < 0.05**                        | At least one group mean differs from others (post-hoc tests needed)         |


Continuous vs Continuous
| Result                            | Interpretation                                            |
| --------------------------------- | --------------------------------------------------------- |
| **Pearson r ≈ +1**                | Strong positive linear relationship                       |
| **Pearson r ≈ -1**                | Strong negative linear relationship                       |
| **Pearson r ≈ 0**                 | No linear relationship (but may have non-linear!)         |
| **Spearman ρ ≈ 1**                | Strong monotonic relationship (rank-based)                |
| **Kendall τ ≈ 1**                 | Strong ordinal association                                |
| **Correlation p-value < 0.05**    | Correlation is statistically significant                  |
| **Scatter points form a line**    | Strong linear relationship                                |
| **Scatter points form a curve**   | Non-linear relationship (consider transformations)        |
| **Scatter points form a cloud**   | Weak or no relationship                                   |
| **Heatmap cell ≈ 1 (dark red)**   | Strong positive correlation                               |
| **Heatmap cell ≈ -1 (dark blue)** | Strong negative correlation                               |
| **Heatmap cell ≈ 0 (neutral)**    | No linear correlation                                     |
| **QQ plot points on 45° line**    | Data follows the theoretical distribution (e.g., normal)  |
| **QQ plot S-shape**               | Heavier or lighter tails than reference distribution      |
| **QQ plot concave up (smile)**    | Right-skewed data                                         |
| **QQ plot concave down (frown)**  | Left-skewed data                                          |
| **Regression p-value < 0.05**     | Slope is significantly different from zero                |
| **R² value**                      | Proportion of variance in Y explained by X (0-1 scale)    |
```
